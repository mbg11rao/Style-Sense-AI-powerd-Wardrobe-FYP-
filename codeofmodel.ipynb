```python
TRAINING_DIR = 'images_compressed'
BATCH_SIZE = 32
IMAGE_SIZE = (224, 224)
EPOCHS = 100
LEARNING_RATE = 0.001
```

```python
import PIL
from pathlib import Path
from PIL import UnidentifiedImageError

path = Path(TRAINING_DIR).rglob("*.jpg")
for img_p in path:
    try:
        img = PIL.Image.open(img_p)
    except PIL.UnidentifiedImageError:
            print(img_p)
```

```python
data = pd.read_csv('images.csv')
data['image'] = data['image']+'.jpg'
data_1 = data[['image', 'label']]
corrupted = ['b72ed5cd-9f5f-49a7-b12e-63a078212a17.jpg', '1d0129a1-f29a-4a3f-b103-f651176183eb.jpg', 'c60e486d-10ed-4f64-abab-5bb698c736dd.jpg', 
            'd028580f-9a98-4fb5-a6c9-5dc362ad3f09.jpg', '040d73b7-21b5-4cf2-84fc-e1a80231b202.jpg', '784d67d4-b95e-4abb-baf7-8024f18dc3c8.jpg']


data_1['image']
data_1 = data_1.drop(data_1[data_1['image'].isin(corrupted)].index, axis=0)
class_names = list(data_1['label'].unique())
print(len(class_names))
class_names

class_dict = dict(zip(class_names, range(len(class_names))))
class_dict
labels_to_remove = ['Skip', 'Not sure', 'Other', 'Blouse']
data_1 = data_1[~data_1['label'].isin(labels_to_remove)]
y_train = data_1['label']
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  rotation_range=40,
                                  width_shift_range=0.3,
                                  height_shift_range=0.3,
                                  zoom_range=0.3,
                                  horizontal_flip=True,
                                  validation_split=0.1)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=data_1,
    directory=TRAINING_DIR,
    x_col='image',
    y_col='label',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=BATCH_SIZE,
    shuffle=True,
    subset='training'
)

validation_generator = train_datagen.flow_from_dataframe(
    dataframe=data_1,
    directory=TRAINING_DIR,
    x_col='image',
    y_col='label',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=BATCH_SIZE,
    shuffle=False,
    subset='validation'
)
```

```python
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2

pre_trained_model = MobileNetV2(input_shape=(224, 224, 3),
                                include_top=False,
                                weights='imagenet')

for layer in pre_trained_model.layers:
    layer.trainable = True

# pre_trained_model.summary()

last_layer = pre_trained_model.get_layer('out_relu')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
x = Flatten()(last_output)
x = Dropout(0.3)(x)
x = Dense(256, activation='relu', kernel_regularizer='l2')(x)
x = Dropout(0.3)(x)
x = Dense(16, activation='softmax')(x)           

model = Model(pre_trained_model.input, x)
```

```python
saved_model_path = 'mobilenetv2_saved_model.h5'

if os.path.exists(saved_model_path):
    print("Loading the trained model from", saved_model_path)
    model = tf.keras.models.load_model(saved_model_path)
else:
    learn_rate = LEARNING_RATE
    es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
    cb = ModelCheckpoint('mobilenetv2.h5', save_best_only=True)
    lrr = ReduceLROnPlateau(monitor='val_accuracy', factor=.1, patience=5, min_lr=1e-5)

    adam = Adam(learning_rate=learn_rate)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])

    history = model.fit(train_generator, epochs=EPOCHS,
                        validation_data=validation_generator, 
                        callbacks=[es, cb, lrr])

    # Load the best model saved by ModelCheckpoint
    model = tf.keras.models.load_model('mobilenetv2.h5')

    # Optionally, save the model with the new filename if needed
    model.save(saved_model_path)
```

```python
classes = list(train_generator.class_indices.keys())
```

```python
def predict_image_with_labels_and_probs(image_path):
    img = keras_image.load_img(image_path, target_size=(224, 224))
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    predictions = model.predict(img_array)
    top3_indices = np.argsort(predictions[0])[::-1][:3]  # Get indices of top 3 predictions
    top3_labels = [classes[idx] for idx in top3_indices]  # Get labels for top 3 predictions
    top3_probs = [predictions[0][idx] for idx in top3_indices]  # Get probabilities for top 3 predictions

    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted labels:\n1. {top3_labels[0]} - Probability: {top3_probs[0]*100:.2f}%\n2. {top3_labels[1]} - Probability: {top3_probs[1]*100:.2f}%\n3. {top3_labels[2]} - Probability: {top3_probs[2]*100:.2f}%")
    plt.show()
```

```python
def closest_color(rgb):
    min_colors = {}
    for key, name in webcolors.CSS3_HEX_TO_NAMES.items():
        r_c, g_c, b_c = webcolors.hex_to_rgb(key)
        rd = (r_c - rgb[0]) ** 2
        gd = (g_c - rgb[1]) ** 2
        bd = (b_c - rgb[2]) ** 2
        min_colors[(rd + gd + bd)] = name
    return min_colors[min(min_colors.keys())]

def remove_background_and_identify_color(image_path):
    image = cv2.imread(image_path)
    original = image.copy()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    resized = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)

    gray = cv2.cvtColor(resized, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    mask = np.zeros_like(resized)
    cv2.drawContours(mask, contours, -1, color=(255, 255, 255), thickness=cv2.FILLED)

    result = cv2.bitwise_and(resized, mask)
    result[mask == 0] = 255

    pixels = result.reshape((result.shape[0] * result.shape[1], 3))
    pixels = np.array([pixel for pixel in pixels if not np.array_equal(pixel, [255, 255, 255])])

    color_info = []

    if len(pixels) > 0:
        kmeans = KMeans(n_clusters=3)
        kmeans.fit(pixels)
        colors = kmeans.cluster_centers_
        labels = kmeans.labels_

        # Count the frequency of each label
        label_counts = np.bincount(labels)

        for i, color in enumerate(colors):
            rgb_color = [int(c) for c in color]
            color_name = closest_color(rgb_color)
            color_prob = label_counts[i] / len(pixels)
            color_info.append((rgb_color, color_name, color_prob))

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 5, 1)
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 5, 2)
    plt.imshow(result)
    plt.title("Background Removed")
    plt.axis("off")

    for i, (color, name, prob) in enumerate(color_info, start=3):
        plt.subplot(1, 5, i)
        plt.imshow([[color]], aspect='auto')
        plt.title(f"{name}\nRGB: {color}\nProb: {prob*100:.2f}%")
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from skimage.color import rgb2lab, deltaE_cie76
import webcolors
import matplotlib.pyplot as plt

def closest_color(rgb):
    min_distance = float('inf')
    closest_name = None
    input_color = rgb2lab(np.uint8([[rgb]]))[0][0]

    for hex_code, name in webcolors.CSS3_HEX_TO_NAMES.items():
        color_rgb = webcolors.hex_to_rgb(hex_code)
        color_lab = rgb2lab(np.uint8([[color_rgb]]))[0][0]
        distance = deltaE_cie76(input_color, color_lab)

        if distance < min_distance:
            min_distance = distance
            closest_name = name

    return closest_name

def remove_background_and_identify_color(image_path):
    image = cv2.imread(image_path)
    original = image.copy()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    resized = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(resized, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    mask = np.zeros_like(resized)
    cv2.drawContours(mask, contours, -1, color=(255, 255, 255), thickness=cv2.FILLED)

    result = cv2.bitwise_and(resized, mask)
    result[mask == 0] = 255

    pixels = result.reshape((result.shape[0] * result.shape[1], 3))
    pixels = np.array([pixel for pixel in pixels if not np.array_equal(pixel, [255, 255, 255])])

    color_info = []

    if len(pixels) > 0:
        kmeans = KMeans(n_clusters=3)
        kmeans.fit(pixels)
        colors = kmeans.cluster_centers_
        labels = kmeans.labels_

        label_counts = np.bincount(labels)

        for i, color in enumerate(colors):
            rgb_color = [int(c) for c in color]
            color_name = closest_color(rgb_color)
            color_prob = label_counts[i] / len(pixels)
            color_info.append((rgb_color, color_name, color_prob))

    # Sort colors by probability
    color_info.sort(key=lambda x: x[2], reverse=True)

    # Visualization
    plt.figure(figsize=(15, 5))

    plt.subplot(1, len(color_info) + 2, 1)
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, len(color_info) + 2, 2)
    plt.imshow(result)
    plt.title("Background Removed")
    plt.axis("off")

    for i, (color, name, prob) in enumerate(color_info, start=3):
        plt.subplot(1, len(color_info) + 2, i)
        plt.imshow([[color]], aspect='auto')
        plt.title(f"{name}\nRGB: {color}\nProb: {prob*100:.2f}%")
        plt.axis("off")

    plt.tight_layout()
    plt.show()

remove_background_and_identify_color('test2.jpg')
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from skimage.color import rgb2lab, deltaE_cie76
import webcolors
import tensorflow as tf
from tensorflow.keras.preprocessing import image as keras_image
import matplotlib.image as mpimg

# Assuming closest_color function is defined as before

def remove_background_and_identify_color_print(image_path):
    image = cv2.imread(image_path)
    original = image.copy()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    resized = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(resized, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    mask = np.zeros_like(resized)
    cv2.drawContours(mask, contours, -1, color=(255, 255, 255), thickness=cv2.FILLED)

    result = cv2.bitwise_and(resized, mask)
    result[mask == 0] = 255

    pixels = result.reshape((result.shape[0] * result.shape[1], 3))
    pixels = np.array([pixel for pixel in pixels if not np.array_equal(pixel, [255, 255, 255])])

    color_info = []

    if len(pixels) > 0:
        kmeans = KMeans(n_clusters=3)
        kmeans.fit(pixels)
        colors = kmeans.cluster_centers_
        labels = kmeans.labels_

        label_counts = np.bincount(labels)

        for i, color in enumerate(colors):
            rgb_color = [int(c) for c in color]
            color_name = closest_color(rgb_color)
            color_prob = label_counts[i] / len(pixels)
            color_info.append((rgb_color, color_name, color_prob))

    # Sort colors by probability
    color_info.sort(key=lambda x: x[2], reverse=True)

    return result, color_info

def predict_image_with_labels_and_probs_print(processed_image_array):
    # Resize the processed image array to the target size required by the model
    resized_img_array = cv2.resize(processed_image_array, (224, 224))

    # Convert the image array to the format required by the model
    img_array = np.expand_dims(resized_img_array, axis=0)
    img_array = img_array / 255.0

    # Get model predictions
    predictions = model.predict(img_array)
    top3_indices = np.argsort(predictions[0])[::-1][:3]  # Get indices of top 3 predictions
    top3_labels = [classes[idx] for idx in top3_indices]  # Get labels for top 3 predictions
    top3_probs = [predictions[0][idx] for idx in top3_indices]  # Get probabilities for top 3 predictions

    return top3_labels, top3_probs

def integrated_image_processing(image_path):
    processed_image, color_info = remove_background_and_identify_color_print(image_path)
    labels_and_probs = predict_image_with_labels_and_probs_print(processed_image)

    # Print results
    print("Color Information:")
    for color, name, prob in color_info:
        print(f"Color: {color}, Name: {name}, Probability: {prob}")

    print("\nLabels and Probabilities:")
    for label, prob in zip(*labels_and_probs):
        print(f"Label: {label}, Probability: {prob}")

# Example usage

```

```python
def integrated_image_processing_name(image_path):
    processed_image, color_info = remove_background_and_identify_color_print(image_path)
    labels_and_probs = predict_image_with_labels_and_probs_print(processed_image)

    # Select the color with the highest probability
    max_color_prob = max(color_info, key=lambda x: x[2])
    color_name = max_color_prob[1]

    # Select the label with the highest probability
    max_label_prob = max(zip(*labels_and_probs), key=lambda x: x[1])
    label_name = max_label_prob[0]

    # ANSI escape codes for bold and some colors
    BOLD = '\033[1m'
    END = '\033[0m'
    RED = '\033[91m'

    # Print the color and label with the highest probability in bold and red
    print(f"{BOLD}{RED}{color_name.capitalize()} {label_name}{END}")
```
